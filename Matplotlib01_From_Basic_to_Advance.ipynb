### Introduction to Matplotlib

**Matplotlib** is a powerful and widely used **data visualization library** in Python. It helps in creating static, animated, and interactive plots with ease. The most commonly used module in this library is `pyplot`, which mimics MATLABâ€™s plotting interface.

---

### ðŸ”§ Basic Features:

| Feature       | Description                                        |
| ------------- | -------------------------------------------------- |
| 2D Plotting   | Line graphs, bar charts, histograms, scatter, etc. |
| Customization | Titles, labels, legends, colors, styles            |
| Exporting     | Save plots as images (PNG, PDF, SVG, etc.)         |
| Integration   | Works with NumPy, pandas, Jupyter, etc.            |

### âœï¸ Basic Example:


## ðŸ§  How It Works: Big Picture

Matplotlib has two main interfaces:

| Interface      | Description                                | Example                        |
|----------------|--------------------------------------------|--------------------------------|
| `pyplot`       | Simple, MATLAB-like interface (easy to use) | `plt.plot(...)`, `plt.show()` |
| `Object-Oriented` | Advanced control, figure & axes objects   | `fig, ax = plt.subplots()`    |

---

## âœ¨ Basic Syntax (Pyplot Style)

```python
import matplotlib.pyplot as plt

# X and Y values
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)               # Line plot
plt.title("My Line Graph")   # Title
plt.xlabel("X Axis")         # X label
plt.ylabel("Y Axis")         # Y label
plt.grid(True)               # Show grid
plt.show()                   # Display the plot
```

## ðŸ§± Figure, Axes, and Subplots (Object-Oriented)

```python
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Line Chart")
ax.set_xlabel("X")
ax.set_ylabel("Y")
```

| Component | Meaning                                       |
|-----------|-----------------------------------------------|
| `Figure`  | The entire canvas (can contain multiple plots)|
| `Axes`    | A single plot area                            |
| `Subplot` | Multiple plots in one figure                  |

Draw 6 plot


### âœ… Options for 6 subplots:

1. **2 rows Ã— 3 columns** â†’ `subplot(2, 3, i)` where `i` = 1 to 6
2. **3 rows Ã— 2 columns** â†’ `subplot(3, 2, i)` where `i` = 1 to 6
3. **6 rows Ã— 1 column** â†’ `subplot(6, 1, i)` for vertical stacking
4. **1 row Ã— 6 columns** â†’ `subplot(1, 6, i)` for horizontal layout

---

### âœ… Example (2 rows Ã— 3 columns):

```python
import matplotlib.pyplot as plt

for i in range(1, 7):
    plt.subplot(2, 3, i)  # 2 rows, 3 columns, subplot i
    plt.plot([1, 2, 3], [j*i for j in [1, 2, 3]])
    plt.title(f"Plot {i}")

plt.tight_layout()
plt.show()
```

Here, the argument `(2, 3, i)` means:

* 2 rows
* 3 columns
* `i`th position (from 1 to 6)

Multiple Bar Plots

````python
import matplotlib.pyplot as plt
import numpy as np

barwidth = 0.25

IT = [12, 22, 3, 4, 5]
ECE = [2, 33, 23, 3, 15]
CSE = [2, 11, 23, 8, 9]

br1 = np.arange(len(IT))
br2 = [x + barwidth for x in br1]
br3 = [x + barwidth for x in br2]

plt.bar(br1, IT, color='r', width=barwidth, edgecolor='gray', label='IT')
plt.bar(br2, ECE, color='g', width=barwidth, edgecolor='gray', label='ECE')
plt.bar(br3, CSE, color='b', width=barwidth, edgecolor='gray', label='CSE')

plt.xlabel('Branch')
plt.ylabel('Students')
plt.title('Students in Different Branches')
plt.xticks([r + barwidth for r in range(len(IT))], ['2018', '2019', '2020', '2021', '2022'])
plt.legend()
plt.show()
````

# Scatter Plot

Plotting With Keyword String

````python
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.title("Scatter Plot with Color and Size")
plt.show()
````

Here's the explanation step by step:

---

### ðŸ§® 1. Create a `data` dictionary:

```python
data = {
    'a': np.arange(50),
    'c': np.random.randint(0, 50, 50),
    'd': np.random.randn(50)
}
```

| Key   | Meaning                                                        |
| ----- | -------------------------------------------------------------- |
| `'a'` | Numbers from 0 to 49 (X-axis values)                           |
| `'c'` | 50 random integers from 0 to 49 (used for **color**)           |
| `'d'` | 50 random numbers from normal distribution (used for **size**) |

---

### ðŸ§® 2. Add a new column `'b'`:

```python
data['b'] = data['a'] + 10 * np.random.randn(50)
```

âœ… `'b'` is created by adding noise (random variation) to `'a'`. This makes the Y-values for scatter points.

---

### ðŸ§® 3. Modify `'d'` to make it usable for size:

```python
data['d'] = np.abs(data['d']) * 100
```

âœ… Takes the absolute value (no negative sizes), then multiplies by 100 to scale the point sizes.

---

### ðŸŽ¨ 4. Plot the scatter plot:

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
```

| Parameter   | Meaning                                   |
| ----------- | ----------------------------------------- |
| `'a'`       | X-axis values                             |
| `'b'`       | Y-axis values                             |
| `c='c'`     | Color of each point based on `'c'`        |
| `s='d'`     | Size of each point based on `'d'`         |
| `data=data` | Tells `matplotlib` to use the `data` dict |

---

### ðŸ·ï¸ 5. Labels and title:

```python
plt.xlabel("entry a")
plt.ylabel("entry b")
plt.title("Scatter Plot with Color and Size")
```

Adds axis labels and a title to the plot.

---

### ðŸ“Š 6. Show the plot:

```python
plt.show()
```

---

This will give you a colorful, varied-size **scatter plot** where:

* X = `'a'`
* Y = `'b'`
* Point color = `'c'`
* Point size = `'d'`
