# Exploratory Data Analysis

Exploratory Data Analysis (EDA) is an important step in data science and data analytics as it visualizes data to understand its main features, find patterns and discover how different parts of the data are connected.

## Types of Exploratory Data Analysis

1. **Univariate Analysis**: This type of analysis focuses on a single variable. It involves summarizing and visualizing the distribution of the variable using measures such as mean, median, mode, variance, and standard deviation. Common visualizations include histograms, box plots, and bar charts.

2. **Bivariate Analysis**: This analysis examines the relationship between two variables. It helps to identify correlations and associations. Common techniques include scatter plots, correlation coefficients, and cross-tabulations.

3. **Multivariate Analysis**: This type of analysis involves more than two variables. It helps to understand complex relationships and interactions among multiple variables. Techniques include multiple regression, principal component analysis (PCA), and cluster analysis.

## Common EDA Techniques
- **Data Cleaning**: Handling missing values, removing duplicates, and correcting inconsistencies in the data.
- **Data Transformation**: Normalizing or standardizing data, creating new features, and encoding categorical variables.
- **Visualization**: Using plots and charts to visualize data distributions and relationships.
- **Statistical Summaries**: Calculating descriptive statistics to summarize the main characteristics of the data.
- **Outlier Detection**: Identifying and handling outliers that may skew the analysis.

## Advanced EDA

Exploratory Data Analysis (EDA) is a crucial first step in any data analysis pipeline. It allows us to gain insights, detect outliers, identify patterns and test assumptions before diving into complex machine learning models. While many associate EDA with data visualization, its just as much about understanding the statistical properties of our data.

### 1. Understanding the Basics of Descriptive Statistics



1.1. Mean

The mean is the average of a set of numbers, calculated by summing all values and dividing by the count of values.

1.2. Median

The median is the middle value in a sorted list of numbers. If the list has an even number of observations, the median is the average of the two middle numbers.

1.3. Mode

The mode is the value that appears most frequently in a dataset. A dataset may have one mode, more than one mode, or no mode at all.

1.4. Standard Deviation

Standard deviation measures the amount of variation or dispersion in a set of values. A low standard deviation indicates that the values tend to be close to the mean, while a high standard deviation indicates that the values are spread out over a wider range.

1.5. Variance

Variance is the average of the squared differences from the mean. It provides a measure of how much the data points differ from the mean.



1.6 iQR (Interquartile Range)

The interquartile range (IQR) is a measure of statistical dispersion, or how spread out the values in a dataset are. It is calculated as the difference between the 75th percentile (Q3) and the 25th percentile (Q1) of the data. IQR = Q3 - Q1

1.7 Skewness

Skewness measures the asymmetry of the probability distribution of a real-valued random variable about its mean. A positive skew indicates that the tail on the right side of the distribution is longer or fatter than the left side, while a negative skew indicates that the tail on the left side is longer or fatter than the right side.



1.7 Kurtosis

Kurtosis measures the "tailedness" of the probability distribution of a real-valued random variable. High kurtosis indicates that the data has heavy tails or outliers, while low kurtosis indicates that the data has light tails or lacks outliers.   

### 2. Visualizing Distributions

Visualization is a critical step in EDA, as it helps to identify patterns, trends and anomalies in the data. Selecting the right type of visualization is crucial to gaining meaningful insights.

2.1. Bar Plot

A bar plot is used to display the frequency or count of categorical data. It consists of rectangular bars where the length of each bar represents the value of the category.



2.2. Stacked Bar Graph

A stacked bar graph is a variation of the bar plot that displays the composition of different categories within a single bar. Each bar is divided into segments, with each segment representing a sub-category.



2.3. Histogram

A histogram is used to represent the distribution of a continuous variable. It consists of bars that represent the frequency of data points within specified intervals (bins).



2.4. Box Plot

A box plot (or box-and-whisker plot) is a graphical representation of the distribution of a dataset. It displays the median, quartiles, and potential outliers.



2.5. Violin Plot

A violin plot is a combination of a box plot and a kernel density plot. It provides a visual representation of the distribution of the data, showing both the summary statistics and the density of the data points.



2.6. Pie Chart

A pie chart is a circular graph divided into slices, where each slice represents a proportion of the whole. It is used to visualize the relative sizes of different categories in a dataset.



2.7. Correlation Heatmap

A correlation heatmap is a graphical representation of the correlation matrix, which shows the pairwise correlations between different variables in a dataset. The values are represented using colors, with darker colors indicating stronger correlations.



2.8. Scatter Plot

A scatter plot is used to visualize the relationship between two continuous variables. Each point on the plot represents an observation in the dataset, with its position determined by the values of the two variables.



### 3. Handling Multivariate Data: Feature Interactions

When dealing with multivariate data, its essential to explore interactions between features to uncover deeper insights. Here are some techniques to handle multivariate data:

3.1. Facet Grids

Facet grids allow you to create multiple plots based on the values of one or more categorical variables. This helps to visualize how the relationship between two continuous variables changes across different categories.



3.2. Pair Plots

A pair plot is a grid of scatter plots that shows the pairwise relationships between multiple continuous variables in a dataset. It also includes histograms or density plots along the diagonal to show the distribution of each variable.




## 4. Identifying Outliers and Anomalies

Outliers and anomalies can significantly impact the results of your analysis. Identifying and handling them appropriately is crucial for accurate insights.

4.1. Z-Scores

Z-scores measure how many standard deviations a data point is from the mean. A Z-score greater than 3 or less than -3 is often considered an outlier.

4.2. Isolation Forest and LOF (Local Outlier Factor)

Isolation Forest and Local Outlier Factor (LOF) are advanced algorithms used for anomaly detection in multivariate data. They help identify data points that deviate significantly from the norm.

## 5. Feature Engineering (Transformations and Interactions)

Feature engineering involves creating new features or transforming existing ones to improve the performance of machine learning models. Here are some common techniques:

5.1. Log Transformation

Log transformation is used to reduce skewness in data and stabilize variance. It is particularly useful for positively skewed data.

5.2. Polynomial Features

Polynomial features involve creating new features by raising existing features to a power. This can help capture non-linear relationships between variables.

5.3. Interaction Features

Interaction features are created by combining two or more features to capture their combined effect on the target variable. This can be done through multiplication, addition, or other mathematical operations.

## 6. Dimensionality Reduction

Dimensionality reduction techniques help to reduce the number of features in a dataset while retaining important information. This is particularly useful for high-dimensional data.

6.1. Principal Component Analysis (PCA)

PCA is a technique that transforms the original features into a new set of uncorrelated features called principal components. These components capture the maximum variance in the data.



6.2. t-SNE (t-Distributed Stochastic Neighbor Embedding)

t-SNE is a non-linear dimensionality reduction technique that is particularly effective for visualizing high-dimensional data in two or three dimensions. It preserves local structures in the data, making it useful for identifying clusters and patterns.



6.3. UMAP (Uniform Manifold Approximation and Projection)

UMAP is another non-linear dimensionality reduction technique that is similar to t-SNE but often faster and better at preserving global structures in the data. It is useful for visualizing complex datasets and identifying clusters.
